```python
def normalize(text):
    text = re.sub('\(', '', text)
    text = re.sub('\)', '', text)
    text = re.sub('\'', '', text)
    text = re.sub(',', '', text)
    text = re.sub('\"', '', text)
    text = re.sub('<[^>]*>', '', text)
    return text.lower()
```
