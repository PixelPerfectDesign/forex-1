```python
import csv
from os.path import expanduser
home = expanduser("~")
with open(home + "/forex/data/datasets/monthly_retail_sales/all.csv", "r") as csvfile:
    with open(home + "/forex/data/datasets/monthly_retail_sales/all_out.csv", "w") as outfile:
        reader = csv.DictReader(csvfile)
        writer = csv.writer(outfile, quoting=csv.QUOTE_ALL)
        #writer.writerow(["NAICS", "Type", "Year", "Month", "Value"])
        for line in reader:
            for key in line.keys():
                if key != "Type" and key != "NAICS":
                    if line[key] != "" and line[key] != "(S)" and line[key] != "(NA)":
                        month, year = key.split(" ")
                        writer.writerow([line['NAICS'], line['Type'].replace(",", ""), year, months[month], line[key].replace(",", "")])
```
